# Forced Response with Step Input (and Zero Initial Condition)

We consider the first-order system:

$$
\dot{x}(t) = a x(t) + b u(t),
\quad
y(t) = c x(t) + d u(t).
$$

<br>

## Step Input Definition
The input is a step of magnitude \(U_m\):

$$
u(t) =
\begin{cases}
0, & t < 0, \\
U_m, & t \geq 0.
\end{cases}
$$

<br>

## Forced Response of the State
With zero initial condition \(x(0)=0\):

$$
x(t) = -\frac{b}{a} U_m \left( 1 - e^{a t} \right).
$$

- The term \((1 - e^{a t})\) captures the exponential transient.  
- The coefficient \(-\tfrac{b}{a} U_m\) is the steady-state value.  
- This is the **forced response**.

<br>

## Output Response
$$
y(t) = c x(t) + d u(t).
$$

Substitute:

$$
y(t) = -\frac{c b}{a} U_m (1 - e^{a t}) + d U_m,
$$

$$
= \left( -\frac{c b}{a}(1 - e^{a t}) + d \right) U_m.
$$

<br>

## Steady-State Behavior
For a stable system (\(a < 0\)):

$$
\lim_{t \to \infty} e^{a t} = 0
$$

so:

$$
x_{ss} = -\frac{b}{a} U_m,
\quad
y_{ss} = \left( -\frac{c b}{a} + d \right) U_m.
$$

<br>

## Steady-State Gain
The steady-state gain from input \(u\) to output \(y\) is:

$$
\frac{y_{ss}}{U_m} = -\frac{c b}{a} + d.
$$

# A General Forced Response

We consider the first-order system:

$$
\dot{x}(t) = a x(t) + b u(t),
\quad x(0) = 0,
$$

under a **step input** \(u(t) = U_m\).

![A generic forced response](https://raw.githubusercontent.com/suann124/AIinTeaching/main/_static/images/L5/Screenshot%202025-10-02%20at%204.39.32%E2%80%AFPM.png)

## General Solution with Zero Initial Condition
From the earlier derivation:

$$
x(t) = -\frac{b}{a} U_m \big(1 - e^{a t}\big).
$$

- As \(t \to \infty\), the exponential decays:
  $$
  x_{ss} = -\frac{b}{a} U_m.
  $$

- The trajectory rises from zero (initial condition) toward \(x_{ss}\).  
- The **time constant** is \(1/|a|\).  
- At one time constant, the error is about \(e^{-1} x_{ss}\) (≈ 37% of the final value remains).

<br>

## Example with Non-Zero Initial Condition
Suppose:
- \(a < 0\) (stable),
- initial condition \(x(0) = x_0 = 1\),
- steady-state value:
  $$
  x_{ss} = -\frac{b}{a} U_m = 2.
  $$

Then the complete solution is the sum of:

1. **Free response (due to IC):**
   $$
   x_{\text{free}}(t) = e^{a t} \cdot x_0.
   $$

2. **Forced response (due to input, zero IC):**
   $$
   x_{\text{forced}}(t) = -\frac{b}{a} U_m \big(1 - e^{a t}\big) = 2 \big(1 - e^{a t}\big).
   $$

<br>

## Complete Solution
By superposition:

$$
x(t) = x_{\text{free}}(t) + x_{\text{forced}}(t).
$$

Substitute:

$$
x(t) = e^{a t} \cdot 1 + \big(2 - 2 e^{a t}\big)
= 2 - e^{a t}.
$$

- At \(t=0\): \(x(0) = 1\) (satisfies the IC).  
- As \(t \to \infty\): \(x(t) \to 2\) (steady state).  
- The exponential term governs the speed of convergence.

![Complete solution](https://raw.githubusercontent.com/suann124/AIinTeaching/main/_static/images/L5/Screenshot%202025-10-02%20at%204.39.49%E2%80%AFPM.png)

**Summary:**  
The total response = free response + forced response.  
In this example:
$$
x(t) = 2 - e^{a t},
$$
showing a transition from 1 → 2 with exponential decay.
